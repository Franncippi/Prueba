```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Creamos datos ficticios de tus visualizaciones mensuales
data = {
    'Mes': ['Ago', 'Sep', 'Oct', 'Nov', 'Dic', 'Ene'],
    'Vistas_Totales': [45000, 48000, 52000, 61000, 75000, 92000],
    'Nuevos_Seguidores': [120, 150, 180, 240, 310, 450]
}

# 2. Convertimos los datos a un DataFrame (formato tabla de Pandas)
df = pd.DataFrame(data)

# 3. Creamos un gráfico para ver la tendencia
plt.figure(figsize=(10, 5))
plt.plot(df['Mes'], df['Vistas_Totales'], marker='o', color='red', linestyle='-', linewidth=2)

# Personalización del gráfico
plt.title('Crecimiento de Visualizaciones Mensuales', fontsize=15)
plt.xlabel('Mes del Año')
plt.ylabel('Cantidad de Vistas')
plt.grid(True, linestyle='--', alpha=0.7)

# Mostrar el gráfico
plt.show()

# 4. Mostrar una estadística simple
print(f"¡Felicidades! Tu crecimiento total fue del {((df['Vistas_Totales'].iloc[-1] / df['Vistas_Totales'].iloc[0]) - 1) * 100:.1f}%")
```
